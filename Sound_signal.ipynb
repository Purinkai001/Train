```python
# ESC-50 classification with signal processing
import librosa
import numpy as np
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import librosa.display
import matplotlib.pyplot as plt


# Path to the ESC-50 dataset
dataset_path = "/content/ESC-50-master"
metadata_path = os.path.join(dataset_path, 'meta/esc50.csv')

# Step 1: Read metadata
def load_metadata(metadata_path):
    metadata = pd.read_csv(metadata_path)
    return metadata

# Step 2: Extract features from each audio file
def extract_features(file_name):
    audio, sample_rate = librosa.load(file_name, sr=None)

    # Extract features (MFCCs, Chroma, Spectrogram, etc.)
    mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=13)
    mfccs_mean = np.mean(mfccs.T, axis=0)

    mel_spectrogram = librosa.feature.melspectrogram(y=audio, sr=sample_rate, n_mels=128)
    mel_spectrogram_mean = np.mean(mel_spectrogram.T, axis=0)

    return np.concatenate([mfccs_mean, mel_spectrogram_mean])

# Step 3: Visualize Mel Spectrogram for a sample file
def visualize_mel_spectrogram(file_name):
    audio, sample_rate = librosa.load(file_name, sr=None)
    mel_spectrogram = librosa.feature.melspectrogram(y=audio, sr=sample_rate, n_mels=128, fmax=8000)
    mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)

    plt.figure(figsize=(10, 6))
    librosa.display.specshow(mel_spectrogram_db, sr=sample_rate, x_axis='time', y_axis='mel', fmax=8000)
    plt.colorbar(format='%+2.0f dB')
    plt.title('Mel Spectrogram')
    plt.tight_layout()
    plt.show()

# Step 4: Load dataset and extract features
def load_dataset(metadata, dataset_path):
    features_list = []
    labels = []

    for index, row in metadata.iterrows():
        file_path = os.path.join(dataset_path, 'audio', row["filename"])
        class_label = row["category"]
        try:
            features = extract_features(file_path)
            features_list.append(features)
            labels.append(class_label)
        except Exception as e:
            print(f"Error extracting features from {file_path}: {e}")

    return np.array(features_list), np.array(labels)

# Step 5: Train the Model
def train_model(features, labels):
    X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

    model = RandomForestClassifier()
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    print(classification_report(y_test, y_pred))

    return model

# Load metadata
metadata = load_metadata(metadata_path)

# Load dataset and extract features
features, labels = load_dataset(metadata, dataset_path)

# Train the model
model = train_model(features, labels)

# Visualize Mel Spectrogram for a sample audio file
#sample_audio_file = os.path.join(dataset_path, 'audio/1-100032-A-0.wav')  # Replace with an actual file from the dataset
visualize_mel_spectrogram('/content/ESC-50-master/audio/1-100032-A-0.wav')
```

```python
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder

def load_spectrogram_dataset(metadata, dataset_path):
    spectrograms = []
    labels = []

    for index, row in metadata.iterrows():
        file_path = os.path.join(dataset_path, 'audio', row["filename"])
        class_label = row["category"]
        try:
            spectrogram = extract_mel_spectrogram(file_path)
            spectrograms.append(spectrogram)
            labels.append(class_label)  # Keep as string for now
        except Exception as e:
            print(f"Error extracting spectrogram from {file_path}: {e}")

    # Reshape data for CNN: (samples, height, width, channels)
    spectrograms = np.array(spectrograms)
    spectrograms = spectrograms[..., np.newaxis]
    label_encoder = LabelEncoder()
    labels_encoded = label_encoder.fit_transform(labels)

    return spectrograms, labels_encoded, label_encoder



def extract_mel_spectrogram(file_name):
    audio, sample_rate = librosa.load(file_name, sr=None)
    mel_spectrogram = librosa.feature.melspectrogram(y=audio, sr=sample_rate, n_mels=128, fmax=8000)
    mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)

    # Resize to a fixed shape (e.g., 128x128) to feed into CNN
    mel_spectrogram_resized = librosa.util.fix_length(mel_spectrogram_db, size=128, axis=1)

    return mel_spectrogram_resized

def build_cnn_model(input_shape):
    model = tf.keras.Sequential([
        tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(50, activation='softmax')  # 50 output classes for ESC-50 dataset
    ])

    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Step 5: Train CNN Model
def train_cnn_model(spectrograms, labels):
    X_train, X_test, y_train, y_test = train_test_split(spectrograms, labels, test_size=0.2, random_state=42)

    input_shape = X_train.shape[1:]  # Shape of input for CNN (128, 128, 1)

    model = build_cnn_model(input_shape)
    model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))

    test_loss, test_acc = model.evaluate(X_test, y_test)
    print(f"Test Accuracy: {test_acc}")

    return model

spectrograms, labels_encoded, label_encoder = load_spectrogram_dataset(metadata, dataset_path)

# Train CNN model
cnn_model = train_cnn_model(spectrograms, labels_encoded)
#Overf
```

```python


# Step 4: Build CNN Model with Dropout, Batch Normalization, and L2 Regularization
def build_new_cnn_model(input_shape):
    model = tf.keras.Sequential([
        tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape,
                               kernel_regularizer=tf.keras.regularizers.l2(0.001)),  # L2 regularization
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Dropout(0.3),  # Dropout layer with 30% dropout

        tf.keras.layers.Conv2D(64, (3, 3), activation='relu',
                               kernel_regularizer=tf.keras.regularizers.l2(0.001)),  # L2 regularization
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Dropout(0.3),

        tf.keras.layers.Conv2D(128, (3, 3), activation='relu',
                               kernel_regularizer=tf.keras.regularizers.l2(0.001)),  # L2 regularization
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Dropout(0.3),

        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(128, activation='relu',
                              kernel_regularizer=tf.keras.regularizers.l2(0.001)),  # L2 regularization
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.4),

        tf.keras.layers.Dense(50, activation='softmax')  # 50 output classes for ESC-50 dataset
    ])

    # Compile the model
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    return model

# Step 5: Implement Early Stopping and Model Training
def train_new_cnn_model(spectrograms, labels):
    X_train, X_test, y_train, y_test = train_test_split(spectrograms, labels, test_size=0.2, random_state=42)

    input_shape = X_train.shape[1:]  # Shape of input for CNN (128, 128, 1)

    model = build_cnn_model(input_shape)

    # Early stopping callback
    early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

    # Train the model with early stopping
    history = model.fit(X_train, y_train, epochs=50, validation_data=(X_test, y_test), callbacks=[early_stopping], batch_size=32)

    # Evaluate the model
    test_loss, test_acc = model.evaluate(X_test, y_test)
    print(f"Test Accuracy: {test_acc}")

    return model, history

# Train the model with reduced overfitting
cnn_model, cnn_history = train_cnn_model(spectrograms, labels_encoded)
```
